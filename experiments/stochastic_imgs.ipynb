```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
sigma = [0.2 * i for i in range(5)]
stochastic = np.array([3893, 3681, 3842, 3608, 3704])
static = np.ceil(stochastic + np.random.sample(len(stochastic)) * stochastic)

plt.scatter(sigma, static, label='Обычный')
plt.scatter(sigma, stochastic, label='С коэффициентами\nдоверия')
plt.title('Сравнение эффективности классического менеджера\n и менеджера с коэффициентами доверия')
plt.xlabel("Значение параметра sigma")
plt.ylabel("Время выполнения работ в плане")
plt.legend()
plt.savefig('sigma')
```
